```python
import pandas as pd
import os
import pathlib as pl
```

```python
data_dir = pl.Path().home() / "Downloads" / "python" / "ottoman" / "data"
print(data_dir)
```

```python
df = pd.read_csv(data_dir / "ottoman.csv",sep=";")
```

```python
mean1 = df.mean()[['taht_suresi']]
mean2 = df.mean()[['taht_cikis_yas']]
mean3 = df.mean()[['taht_inis_yas']]
mean4 = df.mean()[['olum_yas']]
print(f"${mean1} + ${mean2} + ${mean3} + ${mean4}")
```

```python
import matplotlib.pyplot as plt
```

```python
x1 = df["sultan"]
y1 = df["taht_suresi"]
y2 = df["olum_yas"]
# plotting the line 1 points 
plt.plot(x1, y1, label = "line 1")
plt.plot(x1, y2, label = "line 2")
plt.xticks(rotation=90)
plt.figure(figsize=(10,10))
#plt.show()
#plt.savefig('fig1.png', dpi = 100)
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
  
# multiple line plot
plt.xticks(rotation=90)

plt.plot( 'sultan', 'taht_cikis_yas', data=df, marker='o', markerfacecolor='blue', markersize=1, color='skyblue', linewidth=1,label="tahta_cikis_yas")
plt.plot( 'sultan', 'olum_yas', data=df, marker='', color='green', linewidth=2, label="olum_yas")
plt.plot( 'sultan', 'taht_suresi', data=df, marker='', color='red', linewidth=2, linestyle='solid', label="taht_suresi")
plt.legend()
plt.figure(figsize=(20,10)) 
```

```python
df["omur_vs_taht_suresi"] = df["taht_suresi"] / df["olum_yas"] * 100
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
  
# multiple line plot
plt.xticks(rotation=90)
plt.plot( 'sultan', 'taht_suresi', data=df, marker='', color='green', linewidth=2, linestyle='solid', label="taht_suresi")
plt.plot( 'sultan', 'olum_yas', data=df, marker='', color='red', linewidth=2, linestyle='solid', label="vefat_yas")
plt.plot( 'sultan', 'omur_vs_taht_suresi', data=df, marker='', color='blue', linewidth=2, linestyle='dashed', label="omur_vs_hukum")
plt.legend(loc = 'best', prop={'size': 6}, ncol=len(df.columns))
#plt.figure(figsize=(20,10)) 
plt.gcf().subplots_adjust(bottom=0.40)
plt.savefig('fig2.png', dpi = 600)
```
